```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

df_union["FECHA"] = pd.to_datetime(df_union["FECHA"])
df_union["MES"] = df_union["FECHA"].dt.strftime("%Y-%m")

productos_interes = ["MAIZ", "AZUCAR", "PAPA", "TOMATE", "ACEITE"]
df_filtrado = df_union[df_union["PRIMERA_PALABRA"].isin(productos_interes)]

df_agg = df_filtrado.groupby(["MES", "PRIMERA_PALABRA"]).agg({"TOTAL": "sum", "PRECIOREAL": "mean"}).reset_index()

num_productos = len(productos_interes)
nrows = (num_productos + 1) // 2
ncols = 2 if num_productos > 1 else 1

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16, nrows * 6))
axes = axes.flatten()

for i, producto in enumerate(productos_interes):
    data = df_agg[df_agg["PRIMERA_PALABRA"] == producto]

    if len(data) > 1:
        corr, _ = pearsonr(data["PRECIOREAL"], data["TOTAL"])  # Invertido para reflejar el nuevo eje X
        corr_text = f"r = {corr:.2f}"
    else:
        corr_text = "Datos insuficientes"

    sns.regplot(data=data, x="PRECIOREAL", y="TOTAL", scatter=True, ax=axes[i],
                line_kws={"color": "red"}, scatter_kws={"alpha": 0.7, "s": 50})

    axes[i].set_title(f"Relación PRECIOREAL vs TOTAL ({producto})\n{corr_text}")
    axes[i].set_xlabel("Precio Promedio (Miles)")
    axes[i].set_ylabel("Ventas Totales (Millones)")
    axes[i].grid(True)

for j in range(num_productos, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
